```julia
datasir1=readdlm("wt_titr_image_digit.csv",',')
cdatasir1=datasir1[3:end,:]
ydatasir1=cdatasir1[:,2:2:end]
xdatasir1=cdatasir1[:,1:2:end]
ydatasir1[ydatasir1.==""].=0.0
xdatasir1[xdatasir1.==""].=0.0
baseline=copy(ydatasir1)
for i in 1:17
    baseline[:,i].=ydatasir1[1,i]
end
cydatasir1=ydatasir1-baseline
#cydatasir1[109:end,17].=0.0
cydatasir1[cydatasir1.<0.0].=0.0
xdatasir1=xdatasir1.+0.0
intervals=zeros(size(cydatasir1))
for i in 1:17
    for j in 1:size(cydatasir1)[1]-1
        intervals[j,i]=xdatasir1[j+1,i]-xdatasir1[j,i]
        if intervals[j,i]<0.0 
            intervals[j,i]=intervals[j-1,i]
        end
    end
end
norm=zeros(17)
for i in 1:17
    for j in 2:size(cdatasir1)[1]
        norm[i]=norm[i]+cydatasir1[j,i]*intervals[j-1,i]
    end
end
cyndatasir1=copy(cydatasir1)
for i in 1:17
    for j in 1:size(cydatasir1)[1]
        cyndatasir1[j,i]=cydatasir1[j,i]/norm[i]
    end
end
BC=zeros(17)
ff=zeros(17)
for j in 1:size(cyndatasir1)[2]
    firstm=0.0
    secondm=0.0
    thirdm=0.0
    fourthm=0.0
    for i in 2:size(cyndatasir1)[1]
        firstm=firstm+xdatasir1[i,j]*cyndatasir1[i,j]*intervals[i-1,j]
    end
    for i in 2:size(cyndatasir1)[1]
        secondm=secondm+(xdatasir1[i,j]-firstm)^2*cyndatasir1[i,j]*intervals[i-1,j]
        thirdm=thirdm+(xdatasir1[i,j]-firstm)^3*cyndatasir1[i,j]*intervals[i-1,j]
        fourthm=fourthm+(xdatasir1[i,j]-firstm)^4*cyndatasir1[i,j]*intervals[i-1,j]
    end
    skewness=thirdm/secondm^(3/2.0)
    exckurt=fourthm/secondm^2
    BC[j]=(skewness^2+1)/(exckurt)
    ff[j]=firstm
end
BCwt=copy(BC)
datasir1=readdlm("rap1_titr_image_digit.csv",',')
cdatasir1=datasir1[3:end,:]
ydatasir1=cdatasir1[:,2:2:end]
xdatasir1=cdatasir1[:,1:2:end]
ydatasir1[ydatasir1.==""].=0.0
xdatasir1[xdatasir1.==""].=0.0
baseline=copy(ydatasir1)
for i in 1:17
    baseline[:,i].=ydatasir1[1,i]
end
cydatasir1=ydatasir1-baseline
#cydatasir1[109:end,17].=0.0
cydatasir1[cydatasir1.<0.0].=0.0
xdatasir1=xdatasir1.+0.0
intervals=zeros(size(cydatasir1))
for i in 1:17
    for j in 1:size(cydatasir1)[1]-1
        intervals[j,i]=xdatasir1[j+1,i]-xdatasir1[j,i]
        if intervals[j,i]<0.0 
            intervals[j,i]=intervals[j-1,i]
        end
    end
end
norm=zeros(17)
for i in 1:17
    for j in 2:size(cdatasir1)[1]
        norm[i]=norm[i]+cydatasir1[j,i]*intervals[j-1,i]
    end
end
cyndatasir1=copy(cydatasir1)
for i in 1:17
    for j in 1:size(cydatasir1)[1]
        cyndatasir1[j,i]=cydatasir1[j,i]/norm[i]
    end
end
BC=zeros(17)
ff=zeros(17)
for j in 1:size(cyndatasir1)[2]
    firstm=0.0
    secondm=0.0
    thirdm=0.0
    fourthm=0.0
    for i in 2:size(cyndatasir1)[1]
        firstm=firstm+xdatasir1[i,j]*cyndatasir1[i,j]*intervals[i-1,j]
    end
    for i in 2:size(cyndatasir1)[1]
        secondm=secondm+(xdatasir1[i,j]-firstm)^2*cyndatasir1[i,j]*intervals[i-1,j]
        thirdm=thirdm+(xdatasir1[i,j]-firstm)^3*cyndatasir1[i,j]*intervals[i-1,j]
        fourthm=fourthm+(xdatasir1[i,j]-firstm)^4*cyndatasir1[i,j]*intervals[i-1,j]
    end
    skewness=thirdm/secondm^(3/2.0)
    exckurt=fourthm/secondm^2
    BC[j]=(skewness^2+1)/(exckurt)
    ff[j]=firstm
end
BCrap1=copy(BC)
datasir1=readdlm("sir1_titr_image_digit.csv",',')
cdatasir1=datasir1[3:end,:]
ydatasir1=cdatasir1[:,2:2:end]
xdatasir1=cdatasir1[:,1:2:end]
ydatasir1[ydatasir1.==""].=0.0
xdatasir1[xdatasir1.==""].=0.0
baseline=copy(ydatasir1)
for i in 1:17
    baseline[:,i].=ydatasir1[1,i]
end
cydatasir1=ydatasir1-baseline
#cydatasir1[109:end,17].=0.0
cydatasir1[cydatasir1.<0.0].=0.0
xdatasir1=xdatasir1.+0.0
intervals=zeros(size(cydatasir1))
for i in 1:17
    for j in 1:size(cydatasir1)[1]-1
        intervals[j,i]=xdatasir1[j+1,i]-xdatasir1[j,i]
        if intervals[j,i]<0.0 
            intervals[j,i]=intervals[j-1,i]
        end
    end
end
norm=zeros(17)
for i in 1:17
    for j in 2:size(cdatasir1)[1]
        norm[i]=norm[i]+cydatasir1[j,i]*intervals[j-1,i]
    end
end
cyndatasir1=copy(cydatasir1)
for i in 1:17
    for j in 1:size(cydatasir1)[1]
        cyndatasir1[j,i]=cydatasir1[j,i]/norm[i]
    end
end
BC=zeros(17)
ff=zeros(17)
for j in 1:size(cyndatasir1)[2]
    firstm=0.0
    secondm=0.0
    thirdm=0.0
    fourthm=0.0
    for i in 2:size(cyndatasir1)[1]
        firstm=firstm+xdatasir1[i,j]*cyndatasir1[i,j]*intervals[i-1,j]
    end
    for i in 2:size(cyndatasir1)[1]
        secondm=secondm+(xdatasir1[i,j]-firstm)^2*cyndatasir1[i,j]*intervals[i-1,j]
        thirdm=thirdm+(xdatasir1[i,j]-firstm)^3*cyndatasir1[i,j]*intervals[i-1,j]
        fourthm=fourthm+(xdatasir1[i,j]-firstm)^4*cyndatasir1[i,j]*intervals[i-1,j]
    end
    skewness=thirdm/secondm^(3/2.0)
    exckurt=fourthm/secondm^2
    BC[j]=(skewness^2+1)/(exckurt)
    ff[j]=firstm
end
BCsir1=copy(BC)
```
